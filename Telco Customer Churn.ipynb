Perusahaan telekomunikasi **IndoHome** adalah salah satu penyedia layanan internet tetap (fixed broadband) terbesar di Indonesia yang melayani jutaan pelanggan melalui layanan internet, telepon rumah, dan TV interaktif. Di tengah persaingan pasar yang semakin ketat, perusahaan menghadapi tantangan serius dalam mempertahankan pelanggannya.

Pada kuartal terakhir tahun 2021, IndoHome mencatatkan lonjakan tingkat *churn* pelanggan hingga **29,11%**, yang berarti hampir sepertiga dari pelanggan memilih berhenti berlangganan dalam periode tersebut [(ITB, 2021)](https://digilib.itb.ac.id/gdl/view/70721?utm_source=chatgpt.com). Ini menjadi sinyal bahaya karena kehilangan pelanggan tidak hanya berdampak pada penurunan pendapatan, tetapi juga meningkatkan biaya operasional.

Secara global, riset menunjukkan bahwa biaya memperoleh pelanggan baru (*Customer Acquisition Cost* / CAC) bisa mencapai 5 kali lebih tinggi dibanding biaya mempertahankan pelanggan yang sudah ada (*Customer Retention Cost* / CRC) [(Liliyawati, 2024)](https://digilib.unila.ac.id/76438/3/3.%20SKRIPSI%20TANPA%20PEMBAHASAN.pdata). Bahkan, menurut Bain & Company, pengurangan churn sebesar 5% dapat meningkatkan profit hingga 25%-95%, sementara pelanggan yang loyal bisa menghabiskan 67% lebih banyak dibanding pelanggan baru [(Graas.ai, 2024)](https://www.graas.ai/id/blog/dari-listing-menjadi-keuntungan-7-cara-memaksimalkan-pendapatan-dengan-manajemen-marketplace).

Dalam konteks ini, **Chief Marketing Officer (CMO)** menjadi pihak yang paling terdampak dan berkepentingan langsung terhadap strategi penurunan churn. Tanpa pemahaman yang akurat mengenai pola perilaku pelanggan, keputusan alokasi anggaran promosi bisa menjadi tidak efektif.

Oleh karena itu, pendekatan berbasis data seperti *machine learning* menjadi sangat potensial. Dengan memanfaatkan data historis pelanggan, IndoHome dapat mengidentifikasi pelanggan berisiko churn dan memberikan intervensi yang tepat seperti personalisasi layanan atau penawaran diskon, sehingga retensi pelanggan dapat ditingkatkan dan efisiensi biaya pemasaran tercapai.

Tingkat churn sebesar **29,11%** pada akhir tahun 2021 mencerminkan adanya persoalan dalam retensi pelanggan yang dapat berakibat langsung pada kerugian pendapatan dan meningkatnya beban biaya akuisisi pelanggan baru [(ITB, 2021)](https://digilib.itb.ac.id/gdl/view/70721?utm_source=chatgpt.com).

Tanpa pemahaman yang tepat terhadap perilaku pelanggan dan alasan churn, perusahaan berisiko:
- Mengeluarkan biaya promosi kepada pelanggan yang sebenarnya tidak akan churn (*operation loss*),
- Kehilangan pelanggan yang sebenarnya bisa dicegah jika dikenali lebih awal (*profit loss*),
- Dan salah mengambil keputusan pemasaran akibat kurangnya insight berbasis data.

Untuk itu, diperlukan sebuah sistem prediktif yang mampu:
- Memprediksi pelanggan mana yang memiliki potensi tinggi untuk churn,
- Mengidentifikasi fitur-fitur atau layanan yang paling memengaruhi keputusan pelanggan untuk berhenti,
- Memberikan dasar analisis yang kuat untuk strategi retensi yang lebih personal dan efisien.

**Pertanyaan utama yang ingin dijawab:**
1. Apa saja faktor utama yang memengaruhi churn pelanggan di IndoHome?
2. Bagaimana cara memprediksi pelanggan yang berpotensi churn secara akurat menggunakan machine learning?
3. Bagaimana strategi bisnis yang tepat untuk menurunkan churn berdasarkan hasil prediksi dan analisis?

### **1. Estimasi Dampak Finansial (Tanpa Machine Learning)**

---

Sebelum membangun model prediksi churn, dilakukan simulasi dua skenario:

##### Asumsi:

- Total pelanggan: 7.043  
- Churn rate rata-rata dari aktual dataset: 25,5%   
- Biaya promosi retensi per pelanggan: Rp300.000  
  ([Liliyawati, 2024](https://digilib.unila.ac.id/76438/3/3.%20SKRIPSI%20TANPA%20PEMBAHASAN.pdata); [Graas.ai, 2024](https://www.graas.ai/id/blog/dari-listing-menjadi-keuntungan-dengan-manajemen-marketplace))  
- Kerugian akibat kehilangan pelanggan (churn): Rp2.500.000  
  (Estimasi dari ARPU IndiHome ± Rp249.000 × 10 bulan)  
  ([Telkom Indonesia, 2021](https://www.telkom.co.id); [Graas.ai, 2024](https://www.graas.ai/id/blog/dari-listing-menjadi-keuntungan-dengan-manajemen-marketplace))

---

##### **Skenario 1 – Semua Pelanggan Diberi Promosi**
- Seluruh 7.043 pelanggan diberi promosi
- Biaya promosi:  
  $$ 7.043 \times Rp300.000 = Rp2.112.900.000 $$

**Total kerugian: Rp2.112.900.000**

---

##### **Skenario 2 – Tidak Ada Pelanggan Diberi Promosi**
- Tidak ada promosi yang diberikan
- Jumlah pelanggan churn:  
  $$ 7.043 \times 26,5\% \approx 1.867 \text{ pelanggan} $$
- Kerugian akibat churn:  
  $$ 1.867 \times Rp2.500.000 = Rp4.667.500.000 $$

**Total kerugian: Rp5.125.000.000**

---

##### **Ringkasan Skenario**

| Skenario                              | Biaya Promosi     | Kehilangan Profit  | Total Kerugian      |
|---------------------------------------|--------------------|---------------------|----------------------|
| Semua diberi promosi                 | Rp2.112.900.000    | Rp0                 | Rp2.112.900.000      |

### **2. Metric Evaluation yang Digunakan**

---

Prediksi churn pada industri telekomunikasi memiliki dua karakteristik penting:
- **Data tidak seimbang:** proporsi pelanggan churn lebih sedikit (~26%)
- **Kerugian False Negative jauh lebih besar** dibanding False Positive

> Artinya, **menyelamatkan pelanggan churn lebih penting daripada menghindari promosi yang salah sasaran.**

---

##### **Confusion Matrix**
| **Predicted / Actual** | **Churn (1)** | **Not Churn (0)** |
|------------------------|---------------|-------------------|
| **Churn (1)**          | TP: churn terdeteksi dengan benar | FP: pelanggan loyal yang salah diprediksi churn |
| **Not Churn (0)**      | FN: churn tidak terdeteksi        | TN: pelanggan loyal yang diprediksi dengan benar |

---

##### **Metrik yang Digunakan:**

1. **Recall**  
   - Mengukur seberapa banyak pelanggan churn yang berhasil ditangkap model  
   - Prioritas utama, karena FN = kehilangan pelanggan

2. **Precision**  
   - Mengukur seberapa akurat prediksi churn terhadap realitas  
   - Penting untuk efisiensi biaya promosi

3. **F2-Score**  
   - Kombinasi Recall dan Precision, dengan bobot lebih besar pada Recall  
   - Cocok untuk bisnis subscription seperti IndoHome

4. **PRC-AUC (Precision-Recall AUC)**  
   - Memberikan gambaran menyeluruh performa model pada kelas churn  

---

Untuk kasus churn IndoHome:
- **F2-Score dan Recall dijadikan metrik utama**
- **Precision** digunakan sebagai metrik pendamping untuk efisiensi biaya
- **PRC-AUC** sangat penting untuk membandingkan model

Evaluasi ini akan memastikan bahwa model prediktif churn yang dibangun selaras dengan **tujuan bisnis IndoHome**, yaitu **meminimalkan kehilangan pelanggan bernilai tinggi dengan biaya retensi yang efisien**.

Rangkaian analisis akan dilakukan melalui tahapan berikut:

**1. Data Understanding & Exploration** 

---

Tahap awal akan difokuskan pada eksplorasi menyeluruh terhadap dataset pelanggan Telco IndoHome. Analisis univariat dan bivariat akan dilakukan untuk mengenali distribusi fitur, segmentasi pelanggan, serta potensi hubungan antara fitur dengan target churn.

Beberapa metode eksplorasi yang akan digunakan antara lain:

- Visualisasi histogram dan boxplot untuk fitur numerik (tenure, MonthlyCharges, TotalCharges),

- Analisis proporsi churn pada setiap kategori (Contract, InternetService, PaymentMethod) menggunakan count plot dan stacked barplot,

- Korelasi numerik dengan heatmap serta pivot table multivariat untuk mendeteksi segmen pelanggan yang berisiko tinggi,

- Penanganan missing value dan data tidak valid akan diidentifikasi sejak awal.

Tujuan tahap ini adalah mendapatkan insight awal untuk mendukung proses feature engineering dan pemodelan yang tepat sasaran.

**2. Data Cleaning & Preprocessing** 

---

Setelah eksplorasi awal, proses cleaning akan dilakukan untuk memastikan data dalam kondisi siap model. Tahapan ini akan mencakup:

- Penanganan missing values,

- Normalisasi fitur numerik dengan StandardScaler,

- Encoding fitur kategorikal dengan OneHotEncoder,

- Penanganan kategori duplikatif seperti “No internet service”,

- Automasi proses dengan pipeline menggunakan ColumnTransformer.

Fitur customerID akan dihapus karena bersifat identifikasi semata. Dataset akan disiapkan dalam dua bentuk:

- Satu versi untuk eksplorasi (`df_analisis`),

- Satu versi untuk pemodelan (`df_model`).

**3. Feature Engineering & Selection**  

---

Untuk meningkatkan performa model, akan dilakukan rekayasa fitur tambahan seperti:

- `TotalServices`: jumlah layanan digital yang digunakan pelanggan,

- `TenureGroup`: segmentasi durasi langganan (baru, menengah, lama).

Uji coba akan dilakukan terhadap fitur interaksi (PolynomialFeatures) untuk mendeteksi pengaruh kombinasi antar fitur, serta seleksi fitur otomatis menggunakan Recursive Feature Elimination (RFE) dengan model estimator. Hasil seleksi akan diuji berdasarkan performa metrik klasifikasi.

**4. Handling Imbalanced Data**  

---

Karena churn merupakan kelas minoritas (~26.5%), akan dilakukan strategi penanganan ketidakseimbangan (imbalance) agar model tidak bias terhadap kelas mayoritas. Strategi yang akan diuji antara lain:

- SMOTE dan ADASYN untuk oversampling kelas minoritas,

- Analisis metrik dan confusion matrix pada tiap strategi,

- Penyesuaian threshold prediksi untuk meminimalkan false negative.

Model akan dievaluasi dengan metrik berbasis recall dan F2-score, karena mendeteksi churn lebih penting daripada akurasi keseluruhan.


**5. Model Training & Evaluation**  

---

Beberapa algoritma akan diuji dan dibandingkan performanya, antara lain:

1. Logistic Regression,

2. Random Forest,

3. XGBoost dan LightGBM,

4. Decision Tree.

Evaluasi model akan dilakukan menggunakan:

- F2-score sebagai metrik utama,

- Confusion Matrix dan Classification Report,

- Precision-Recall Curve (PRC) dan ROC Curve,

- Analisis threshold (cut-off) untuk menentukan titik optimal antara recall dan precision.

Model terbaik akan dioptimalkan melalui RandomizedSearchCV, dengan pertimbangan efisiensi dan kemampuan interpretasi.

**6. Deployment & Business Strategy**  

---

Model terbaik akan dikemas dalam pipeline lengkap (preprocessing + model + resampling) dan di-deploy menggunakan Streamlit WebApp yang dapat digunakan oleh CMO dan tim CRM IndoHome. Aplikasi ini akan mendukung:

- Prediksi churn individu maupun batch (CSV),

- Visualisasi status risiko dan rekomendasi tindakan,

- Interpretasi hasil model secara langsung menggunakan fitur koefisien Logistic Regression.

Selain itu, model akan dianalisis dari sudut explainability untuk memastikan hasil yang dapat dipertanggungjawabkan secara bisnis. Strategi retensi berbasis data akan disusun berdasarkan fitur-fitur utama yang berkontribusi terhadap churn, dan hasil prediksi akan digunakan untuk simulasi finansial sebagai bukti dampak ekonomi model ini.

Dengan pendekatan ini, proyek churn IndoHome tidak hanya akan menghasilkan prediksi, tetapi juga mendorong transformasi pengambilan keputusan marketing menjadi lebih efisien, tepat sasaran, dan data-driven.

Berdasarkan dataset **Telco Customer Churn** dari Kaggle, variabel target `Churn` didefinisikan secara spesifik sebagai berikut:

> Pelanggan dikatakan mengalami **"Churn" (`Churn = Yes`)** jika pelanggan tersebut **telah secara eksplisit meninggalkan layanan (berhenti berlangganan atau kontrak tidak diperpanjang)** dalam kurun waktu **1 bulan terakhir**.

Beberapa kondisi spesifik yang menyebabkan seorang pelanggan dilabeli churn dalam dataset ini:

- Pelanggan tidak memperpanjang atau memperbaharui kontrak layanan mereka setelah berakhir masa kontrak (misalnya kontrak bulanan selesai dan tidak diperpanjang).
- Pelanggan secara eksplisit membatalkan layanan (mengajukan penghentian layanan).
- Pelanggan tidak lagi melakukan pembayaran sehingga layanan dihentikan secara paksa oleh perusahaan.

Sebaliknya, pelanggan yang masih menggunakan layanan aktif dan masih berada dalam status pembayaran/kontrak aktif dilabeli **"Tidak Churn" (`Churn = No`)**.

#### **1. Univariate Analysis - Fitur Numerik**

Analisis univariat dilakukan pada tiga fitur numerik utama dalam dataset pelanggan IndoHome: `tenure`, `MonthlyCharges`, dan `TotalCharges`. Tujuannya adalah untuk memahami distribusi, mendeteksi pola umum, serta melihat keberadaan outlier atau anomali sebelum masuk ke tahap pemodelan.

```python
# Plot boxplot untuk tenure, MonthlyCharges, TotalCharges
fig, axes = plt.subplots(1, 3, figsize=(15,5))
for ax, col in zip(axes, numerical_features):
    sns.boxplot(y=data[col], ax=ax, color='lightgreen')
    ax.set_title(f'{col} Boxplot')
plt.tight_layout()
plt.show()
```

```python
# Plot QQ-plot untuk tenure, MonthlyCharges, TotalCharges
fig, axes = plt.subplots(1, 3, figsize=(15,5))
for ax, col in zip(axes, numerical_features):
    stats.probplot(data[col], dist="norm", plot=ax)
    ax.set_title(f'QQ Plot of {col}')
plt.tight_layout()
plt.show()
```

2. **MonthlyCharges**

#### **1. Bivariate Analysis - Fitur Numerik vs Churn**

Untuk variabel numerik, kita menggunakan plot perbandingan distribusi (misal boxplot atau violin plot) antara pelanggan churn Yes vs No. Di sini digunakan violin plot agar dapat melihat distribusi lengkap berikut median dan IQR di dalamnya. Kita plot untuk tiga variabel numerik: `tenure`, `MonthlyCharges`, dan `TotalCharges` terhadap dua kategori churn.

```python
# Violin plot: distribusi numerik vs Churn
fig, axes = plt.subplots(3, 1, figsize=(6, 15))
for ax, col in zip(axes, numerical_features):
    sns.violinplot(x='Churn', y=col, data=data, ax=ax, palette='Set1', inner='quartile')
    ax.set_title(f'{col} vs Churn')
plt.tight_layout()
plt.show()
```

Distribusi fitur numerik berdasarkan churn (violin plot). Plot di atas membandingkan distribusi tenure, MonthlyCharges, dan TotalCharges untuk pelanggan churn (Yes, warna biru) vs tidak churn (No, warna merah). Terlihat perbedaan yang sangat mencolok:

1.  `Tenure` vs `Churn`: Pelanggan yang churn cenderung memiliki tenure yang jauh lebih rendah. Violin plot churn Yes (biru) sangat terkonsentrasi di bagian bawah (median ~10 bulan, banyak yang hanya berlangganan di bawah 2 tahun), sedangkan pelanggan yang tidak churn memiliki sebaran tenure hingga 72 bulan dengan median yang jauh lebih tinggi (~38 bulan). Ini mengindikasikan bahwa pelanggan baru atau berlangganan singkat lebih mungkin churn dibanding pelanggan yang sudah lama bertahan.

2. `MonthlyCharges` vs `Churn`: Distribusi biaya bulanan untuk pelanggan churn sedikit lebih tinggi dibanding yang tidak churn. Terlihat median MonthlyCharges untuk churn Yes lebih tinggi (sekitar $80) daripada yang No (~$65). Violin biru (churn) lebih tebal di kisaran nilai tinggi (70-100), sedangkan violin merah lebih tebal di kisaran lebih rendah (20-50). Hal ini menunjukkan pelanggan dengan tagihan bulanan tinggi cenderung lebih sering churn. Mungkin pelanggan dengan paket mahal (banyak layanan) lebih rentan berhenti, bisa jadi karena merasa biaya terlalu tinggi atau tidak sesuai ekspektasi layanan.

  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">4.1 | Menangani Missing Values</h2>

<div style='width: 100%; height: 0.05px; background-color: #BDC3C7; border-radius: 10px; margin-top: 20px; margin-bottom: 20px;'></div>

Misalnya output info awal menunjukkan kolom TotalCharges bertipe object (string) padahal seharusnya numerik, dan jumlah null secara langsung semua 0. Ini memberi tanda bahwa mungkin ada nilai kosong yang tidak terdeteksi sebagai NaN. Dari penjelasan tersebut, kita curiga kolom `TotalCharges` mengandung nilai kosong dalam bentuk string kosong atau spasi (karena tipe data terdeteksi object). Biasanya pada dataset Telco, pelanggan baru (tenure 0 bulan) belum memiliki `TotalCharges` (kosong).

```python
# Membuat duplikasi dataframe
df_cleaned = data.copy()
```

```python
# Hitung jumlah entri 'TotalCharges' yang kosong atau blank
blank_count = (df_cleaned['TotalCharges'] == " ").sum()
print("Jumlah data TotalCharges blank:", blank_count)

# Konversi 'TotalCharges' menjadi numerik (float), ubah blank menjadi NaN terlebih dahulu
data['TotalCharges'] = df_cleaned['TotalCharges'].replace(" ", np.nan)
data['TotalCharges'] = df_cleaned['TotalCharges'].astype(float)

# Setelah konversi, tangani nilai NaN (dari entri blank)
print("Jumlah NA di TotalCharges setelah konversi:", df_cleaned['TotalCharges'].isna().sum())
# Isi NaN dengan 0, karena pelanggan tenure 0 bulan logis TotalCharges=0
df_cleaned['TotalCharges'].fillna(0, inplace=True)

# Verifikasi tidak ada missing di TotalCharges lagi
print("Jumlah NA di TotalCharges setelah imputasi:", df_cleaned['TotalCharges'].isna().sum())
```

Kita mengisi TotalCharges yang kosong dengan 0 karena pelanggan dengan tenure 0 bulan belum memiliki tagihan, jadi wajar total charge-nya 0. Alternatif lain, kita bisa menghapus baris tersebut (karena hanya 11 dari 7043 baris, <0.2%), namun kita memilih mengimputasi 0 karena lebih masuk akal secara bisnis (belum ada penggunaan = tagihan 0) dan agar tidak membuang data pelanggan baru.

>

**Outlier**

---

Kolom tenure, MonthlyCharges, dan TotalCharges merupakan fitur numerik utama yang berpotensi mengandung nilai outlier. Outlier adalah nilai yang jauh dari mayoritas data dan dapat mempengaruhi hasil analisis statistik maupun performa model prediktif. Untuk mengidentifikasi outlier, digunakan metode Interquartile Range (IQR), yang mendeteksi nilai-nilai di luar rentang kuartil bawah (Q1) dan kuartil atas (Q3) sebesar 1.5 kali IQR. Pemeriksaan ini membantu kita memahami distribusi data dan memutuskan apakah perlu menangani outlier sebelum melanjutkan ke tahap modeling.

```python
for col in ['tenure', 'MonthlyCharges', 'TotalCharges']:
    q1 = df_cleaned[col].quantile(0.25)
    q3 = df_cleaned[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers = df_cleaned[(df_cleaned[col] < lower) | (df_cleaned[col] > upper)]
    n_outliers = outliers.shape[0]
    total = df_cleaned.shape[0]
    percentage = n_outliers / total * 100
    print(f"{col}: {n_outliers} outliers ({percentage:.2f}% dari {total} data)")
```

  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">6.1 | Memisahkan Fitur dan Target</h2>

<div style='width: 100%; height: 0.05px; background-color: #BDC3C7; border-radius: 10px; margin-top: 20px; margin-bottom: 20px;'></div>

Langkah pertama adalah memisahkan fitur (`X`) dan target (`y`) dari dataset. Ini memungkinkan kita untuk melakukan preprocessing hanya pada fitur yang akan digunakan dalam model.


```python
# Memisahkan fitur dan target
X = df_model.drop(columns='Churn')
y = df_model['Churn'].map({'No': 0, 'Yes': 1})

print(X.shape)
print(y.shape)
```

<div style="text-align: left;">
  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">6.2 | Membagi Data Training & Testing</h2>
</div>

<div style='width: 100%; height: 0.05px; background-color: #BDC3C7; border-radius: 10px; margin-top: 20px; margin-bottom: 20px;'></div>

Agar model dapat diuji dengan baik, dataset dibagi menjadi 80% data training dan 20% data testing menggunakan `train_test_split` dengan stratifikasi berdasarkan target (`y`). Pembagian data secara terpisah untuk training dan testing penting agar evaluasi model tidak bias. Stratified splitting memastikan distribusi kelas target pada kedua subset mirip.

```python
# Split dataset (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"y_train: {y_train.shape}, y_test: {y_test.shape}")
```

<div style="text-align: left;">
  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">6.3 | Pengecekkan Imbalance Data</h2>
</div>

<div style='width: 100%; height: 0.05px; background-color: #BDC3C7; border-radius: 10px; margin-top: 20px; margin-bottom: 20px;'></div>

Pada tahap ini, kita akan melakukan analisis distribusi kelas target (y_train) untuk mengetahui apakah terdapat ketidakseimbangan data (imbalanced data). Ketidakseimbangan data terjadi jika salah satu kelas jauh lebih dominan dibandingkan kelas lainnya, yang dapat menyebabkan model machine learning menjadi bias terhadap kelas mayoritas.

Mengapa Pengecekan Imbalance Data Penting?
1. Model cenderung mengabaikan kelas minoritas jika dataset sangat tidak seimbang.
2. Akurasi tinggi tidak selalu berarti model bagus, karena model bisa saja hanya memprediksi kelas mayoritas.
3. Jika ditemukan ketidakseimbangan, strategi penanganan seperti oversampling, SMOTE, atau class weight adjustment perlu diterapkan.

```python
imbalance_info = {
    "Distribusi Target y (Train Set)": y_train.value_counts(normalize=True).round(3),
    "Total Data Train": len(y_train),
    "Total Data Test": len(y_test)
}

imbalance_info
```

Distribusi kelas target pada data training menunjukkan bahwa 73.5% pelanggan tidak churn (label = 0) dan hanya 26.5% pelanggan churn (label = 1). Ini mengindikasikan ketidakseimbangan kelas (class imbalance).

Ketidakseimbangan ini berpotensi membuat model bias terhadap kelas mayoritas (tidak churn), sehingga model bisa tampak akurat secara keseluruhan, namun gagal mengenali pelanggan yang berisiko churn.

Diperlukan strategi penanganan imbalance seperti oversampling (misalnya ADASYN atau SMOTE) atau penggunaan class weight untuk memastikan model mampu mengenali kelas minoritas dengan baik. Langkah ini penting agar model prediksi churn dapat digunakan secara efektif dalam strategi retensi pelanggan IndoHome.

<div style="text-align: left;">
  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">6.4 | Pengelompokan Kolom Berdasarkan Kebutuhan Preprocessing</h2>
</div>

<div style='width: 100%; height: 0.05px; background-color: #BDC3C7; border-radius: 10px; margin-top: 20px; margin-bottom: 20px;'></div>

Langkah ini bertujuan untuk mengelompokkan fitur menjadi dua jenis utama sesuai kebutuhan preprocessing:

1. Fitur numerik membutuhkan transformasi seperti scaling (normalisasi/standarisasi),

2. Fitur kategorikal membutuhkan encoding (One-Hot atau Ordinal).

Pengelompokan ini akan menjadi dasar bagi proses transformasi otomatis menggunakan ColumnTransformer.

```python
# Identifikasi fitur numerik dan kategorikal
numeric_features = ['tenure', 'MonthlyCharges', 'TotalCharges']
categorical_features = [col for col in X.columns if col not in numeric_features]
```

<div style="text-align: left;">
  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">6.5 | Generate New Features</h2>
</div>

<div style='width: 100%; height: 0.05px; background-color: #BDC3C7; border-radius: 10px; margin-top: 20px; margin-bottom: 20px;'></div>

Feature engineering dilakukan untuk meningkatkan kualitas input model dengan menambahkan dua fitur baru:

1. `TotalServices`: Jumlah layanan tambahan yang dimiliki pelanggan (dihitung dari fitur-fitur seperti OnlineSecurity, TechSupport, dll).

2. `TenureGroup`: Segmentasi pelanggan berdasarkan durasi langganan ke dalam kategori 'Baru', 'Menengah', dan 'Lama'.

```python
# Feature Engineering (menambahkan ke X_train dan X_test)
# Total layanan tambahan yang dimiliki pelanggan

service_cols = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                'TechSupport', 'StreamingTV', 'StreamingMovies']
X_train['TotalServices'] = (X_train[service_cols] == 'Yes').sum(axis=1)
X_test['TotalServices'] = (X_test[service_cols] == 'Yes').sum(axis=1)
```

```python
# Segmentasi durasi berlangganan (tenure)
def tenure_group(x):
    if x <= 12:
        return 'Baru'
    elif x <= 24:
        return 'Menengah'
    else:
        return 'Lama'
```

```python
X_train['TenureGroup'] = X_train['tenure'].apply(tenure_group)
X_test['TenureGroup'] = X_test['tenure'].apply(tenure_group)
```

```python
# Tambahkan fitur baru ke dalam daftar fitur kategorikal
categorical_features += ['TenureGroup']
numeric_features += ['TotalServices']

# Output informasi sebelum lanjut ke pipeline
{
    "Imbalance Info": imbalance_info,
    "Fitur Numerik": numeric_features,
    "Fitur Kategorikal": categorical_features,
    "Contoh TotalServices": X_train['TotalServices'].value_counts().sort_index(),
    "Contoh TenureGroup": X_train['TenureGroup'].value_counts()
}
```

Fitur tambahan berhasil meningkatkan dimensi informasi dalam data:

1. Mayoritas pelanggan tergolong Lama (>24 bulan),

2. Banyak pelanggan hanya menggunakan 0–2 layanan tambahan, mengindikasikan potensi upselling.

Fitur-fitur baru ini diharapkan dapat meningkatkan performa dan interpretabilitas model prediksi churn.

<div style="text-align: left;">
  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">6.6 | pipeline preprocessing & ColumnTransformer</h2>
</div>

<div style='width: 100%; height: 0.05px; background-color: #BDC3C7; border-radius: 10px; margin-top: 20px; margin-bottom: 20px;'></div>

Pada tahap ini, digunakan ColumnTransformer untuk mengotomatisasi preprocessing sesuai jenis data:

- StandardScaler diterapkan pada fitur numerik (`tenure`, `MonthlyCharges`, `TotalCharges`, `TotalServices`) karena:
  - Rentang skala antar fitur berbeda,
  - Tidak ditemukan outlier ekstrem,
  - Distribusi data relatif wajar,

- OneHotEncoder digunakan untuk fitur kategorikal karena:
  - Sebagian besar fitur bersifat nominal (tanpa urutan logis),
  - Encoding biner menghindari asumsi urutan yang salah,

Pendekatan ini memastikan setiap fitur diproses secara tepat sesuai karakteristiknya, meningkatkan keakuratan dan stabilitas model prediksi churn.

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features)
    ]
)
```

<div style="text-align: left;">
  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">6.7 | Fit dan transformasi data</h2>
</div>

<div style='width: 100%; height: 0.05px; background-color: #BDC3C7; border-radius: 10px; margin-top: 20px; margin-bottom: 20px;'></div>

Semua pipeline digabungkan dalam `ColumnTransformer`, sehingga preprocessing dapat diterapkan secara bersamaan pada berbagai tipe fitur.

```python
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

<div style="text-align: left;">
  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">6.8 | Konversi hasil transformasi ke DataFrame</h2>
</div>

<div style='width: 100%; height: 0.05px; background-color: #BDC3C7; border-radius: 10px; margin-top: 20px; margin-bottom: 20px;'></div>

Agar lebih mudah dianalisis, hasil transformasi dikonversi ke dalam `DataFrame`.

```python
# Ambil nama kolom hasil encoding
ohe = preprocessor.named_transformers_['cat']
encoded_cat_cols = ohe.get_feature_names_out(categorical_features)

# Gabungkan nama kolom numerik + hasil one-hot
all_feature_names = numeric_features + list(encoded_cat_cols)

# Buat DataFrame dari hasil preprocessing
X_train_df = pd.DataFrame(X_train_processed.toarray() if hasattr(X_train_processed, "toarray") else X_train_processed,
                          columns=all_feature_names, index=X_train.index)
X_test_df = pd.DataFrame(X_test_processed.toarray() if hasattr(X_test_processed, "toarray") else X_test_processed,
                         columns=all_feature_names, index=X_test.index)
```

<div style="text-align: left;">
  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">6.9 | Evaluasi Hasil Preprocessing</h2>
</div>

5. `LightGBM`: (Mendukung scale_pos_weight)

Eksperimen ini bertujuan untuk mengevaluasi pengaruh penambahan fitur baru berbentuk polinomial (Polynomial Features) terhadap performa model dalam memprediksi pelanggan churn.

**Tujuan Eksperimen:**
1. Menguji apakah kombinasi interaktif antar fitur numerik (misalnya `tenure` × `MonthlyCharges`) dapat membantu model menemukan pola churn yang sebelumnya tidak terlihat.

2. Menilai apakah penambahan fitur polinomial membantu model linear seperti Logistic Regression, yang secara bawaan tidak bisa menangkap interaksi non-linear.

3. Eksperimen dilakukan tanpa penyesuaian class_weight atau scale_pos_weight, agar kita bisa mengisolasi efek dari fitur baru secara murni.

**Metode yang Digunakan:**
1. Fitur numerik: `tenure`, `MonthlyCharges`, dan `TotalCharges`

2. Penambahan fitur interaktif (degree = 2, interaction_only = True, include_bias = False)

3. Model yang diuji: Logistic Regression (karena sangat dipengaruhi oleh struktur fitur)

4. Evaluasi menggunakan F2-score dan validasi silang stratifikasi (5-fold Stratified KFold)

**Alasan Tidak Menggunakan Polynomial Features untuk Semua Model**
Model berbasis pohon seperti Decision Tree, Random Forest, XGBoost, dan LightGBM secara alami sudah mampu menangkap interaksi antar fitur, sehingga menambahkan polynomial features berpotensi menambah kompleksitas tanpa peningkatan performa signifikan (bahkan bisa overfitting).

**Tidak Menggunakan Class Weight / Scale Pos Weight Dulu**
Eksperimen ini bertujuan untuk menilai murni dampak dari fitur polinomial, jadi kita tidak menggunakan teknik balancing apa pun. Nanti pada 6.3, kita akan menggabungkannya.

```python
# Polynomial Transformation
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_train_poly = poly.fit_transform(X_train_df)
X_test_poly = poly.transform(X_test_df)

# Model definitions
models = {
    "Logistic Regression": LogisticRegression(random_state=42, max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'),
    "LightGBM": LGBMClassifier(random_state=42)
}

# Cross-validation setup
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
f2_scorer = make_scorer(fbeta_score, beta=2)

# Results container
results = []

# Loop through models
for model_name, model in models.items():
    print(f"Evaluating {model_name}...")

    # Baseline
    f2_scores_base = cross_val_score(model, X_train_df, y_train, cv=kf, scoring=f2_scorer)
    base_mean = np.mean(f2_scores_base)
    base_std = np.std(f2_scores_base)
    model.fit(X_train_df, y_train)
    f2_test_base = fbeta_score(y_test, model.predict(X_test_df), beta=2)

    # Polynomial Features
    f2_scores_poly = cross_val_score(model, X_train_poly, y_train, cv=kf, scoring=f2_scorer)
    poly_mean = np.mean(f2_scores_poly)
    poly_std = np.std(f2_scores_poly)
    model.fit(X_train_poly, y_train)
    f2_test_poly = fbeta_score(y_test, model.predict(X_test_poly), beta=2)

    results.append({
        "Model": model_name,
        "Baseline F2 Train Mean": round(base_mean, 4),
        "Baseline F2 Train Std Dev": round(base_std, 4),
        "Baseline F2 Test": round(f2_test_base, 4),
        "Polynomial F2 Train Mean": round(poly_mean, 4),
        "Polynomial F2 Train Std Dev": round(poly_std, 4),
        "Polynomial F2 Test": round(f2_test_poly, 4)
    })

# Convert to DataFrame
df_results_poly = pd.DataFrame(results)
df_results_poly
```

Eksperimen ini bertujuan untuk mengevaluasi apakah penambahan fitur interaksi (Polynomial Features) dapat memberikan peningkatan performa tambahan setelah sebelumnya model telah ditingkatkan menggunakan strategi penanganan data tidak seimbang melalui class_weight="balanced" atau scale_pos_weight.

Tujuan Eksperimen:

1. Melihat apakah penambahan kombinasi fitur hasil interaksi dapat membantu model dalam mengenali pola kompleks yang belum terdeteksi sebelumnya.

2. Menguji apakah strategi penambahan fitur dan balancing dapat berkolaborasi secara sinergis atau justru menimbulkan overfitting dan penurunan performa.

Metodologi:

- Melatih ulang seluruh model dari eksperimen 6.2 menggunakan data hasil transformasi polynomial (degree=2, interaction_only=True).

- Semua model juga tetap menggunakan strategi class balancing seperti sebelumnya (class_weight atau scale_pos_weight).

- Evaluasi dilakukan dengan 5-fold cross-validation pada data training dan perhitungan F2-score pada data test.

```python
# Polynomial Features
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_train_poly = poly.fit_transform(X_train_df)
X_test_poly = poly.transform(X_test_df)

# Scale Pos Weight
scale_pos_weight = y_train.value_counts()[0] / y_train.value_counts()[1]

# Cross-validation & scoring
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
f2_scorer = make_scorer(fbeta_score, beta=2)

models = {
    "Logistic Regression": LogisticRegression(random_state=42, class_weight="balanced", max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42, class_weight="balanced"),
    "Random Forest": RandomForestClassifier(random_state=42, class_weight="balanced"),
    "XGBoost": XGBClassifier(random_state=42, scale_pos_weight=scale_pos_weight, use_label_encoder=False, eval_metric="logloss"),
    "LightGBM": LGBMClassifier(random_state=42, scale_pos_weight=scale_pos_weight)
}

results = []

for model_name, model in models.items():
    print(f"Evaluating {model_name}...")

    # Model dengan Class Weight / Scale Pos Weight
    f2_weighted = cross_val_score(model, X_train_df, y_train, cv=kf, scoring=f2_scorer)
    weighted_mean = np.mean(f2_weighted)
    weighted_std = np.std(f2_weighted)
    model.fit(X_train_df, y_train)
    weighted_test = fbeta_score(y_test, model.predict(X_test_df), beta=2)

    # Model dengan Class Weight / Scale Pos Weight + Polynomial Features
    f2_poly_weighted = cross_val_score(model, X_train_poly, y_train, cv=kf, scoring=f2_scorer)
    poly_weighted_mean = np.mean(f2_poly_weighted)
    poly_weighted_std = np.std(f2_poly_weighted)
    model.fit(X_train_poly, y_train)
    poly_weighted_test = fbeta_score(y_test, model.predict(X_test_poly), beta=2)

    results.append({
        "Model": model_name,
        "Weighted F2 Train Mean": round(weighted_mean, 4),
        "Weighted F2 Train Std Dev": round(weighted_std, 4),
        "Weighted F2 Test": round(weighted_test, 4),
        "Poly Weighted F2 Train Mean": round(poly_weighted_mean, 4),
        "Poly Weighted F2 Train Std Dev": round(poly_weighted_std, 4),
        "Poly Weighted F2 Test": round(poly_weighted_test, 4)
    })

df_results_weighted_poly = pd.DataFrame(results)
display(df_results_weighted_poly)
```

Eksperimen ini bertujuan untuk mengevaluasi apakah penambahan fitur interaksi (polynomial degree 2) dapat meningkatkan performa model yang sudah menerapkan strategi balancing class (class_weight atau scale_pos_weight).

1. Penambahan polynomial features (degree=2, interaction only) tidak memberikan peningkatan yang signifikan untuk model-model yang sudah menerapkan class weight / scale_pos_weight.

Dengan menggunakan RFECV (Recursive Feature Elimination with Cross Validation), kita memilih subset fitur paling relevan berdasarkan pentingnya terhadap target variabel. Pendekatan ini dilakukan dengan estimator Random Forest, karena mampu menangkap interaksi non-linear dan memberikan importance score yang andal.

#### **RFE - RFECV**
---

```python
from sklearn.feature_selection import RFECV

# Gunakan RFECV untuk memilih fitur optimal
selector = RFECV(estimator=RandomForestClassifier(random_state=42), step=1, cv=5, scoring='f1')
X_train_selected = selector.fit_transform(X_train_df, y_train)
X_test_selected = selector.transform(X_test_df)

print(f"Jumlah fitur optimal terpilih: {selector.n_features_}")
```

Setelah fitur terpilih (sebanyak 23 fitur), performa model dibandingkan dengan baseline yang menggunakan seluruh fitur. Ini memungkinkan kita untuk menilai apakah pengurangan dimensi dapat meningkatkan efisiensi model tanpa mengorbankan akurasi.

```python
# Model untuk evaluasi
models = {
    "Logistic Regression": LogisticRegression(random_state=42, class_weight="balanced", max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42, class_weight="balanced"),
    "Random Forest": RandomForestClassifier(random_state=42, class_weight="balanced"),
    "XGBoost": XGBClassifier(random_state=42, scale_pos_weight=scale_pos_weight, use_label_encoder=False, eval_metric='logloss'),
    "LightGBM": LGBMClassifier(random_state=42, scale_pos_weight=scale_pos_weight)
}

results_rfe = []

for model_name, model in models.items():
    print(f"Evaluating {model_name} with RFE...")

    # Baseline (tanpa feature selection)
    f2_base = cross_val_score(model, X_train_df, y_train, cv=kf, scoring=f2_scorer)
    base_mean = np.mean(f2_base)
    base_std = np.std(f2_base)
    model.fit(X_train_df, y_train)
    base_test = fbeta_score(y_test, model.predict(X_test_df), beta=2)

    # Dengan RFE (fitur terpilih)
    f2_rfe = cross_val_score(model, X_train_selected, y_train, cv=kf, scoring=f2_scorer)
    rfe_mean = np.mean(f2_rfe)
    rfe_std = np.std(f2_rfe)
    model.fit(X_train_selected, y_train)
    rfe_test = fbeta_score(y_test, model.predict(X_test_selected), beta=2)

    results_rfe.append({
        "Model": model_name,
        "Baseline F2 Train Mean": round(base_mean, 4),
        "Baseline F2 Train Std Dev": round(base_std, 4),
        "Baseline F2 Test": round(base_test, 4),
        "RFE F2 Train Mean": round(rfe_mean, 4),
        "RFE F2 Train Std Dev": round(rfe_std, 4),
        "RFE F2 Test": round(rfe_test, 4)
    })

df_results_rfe = pd.DataFrame(results_rfe)
display(df_results_rfe)
```

Eksperimen ini bertujuan mengevaluasi apakah seleksi fitur menggunakan Recursive Feature Elimination (RFE) dapat meningkatkan performa model dengan memilih hanya fitur yang paling relevan. Berdasarkan hasil RFECV, sebanyak 23 fitur optimal berhasil dipilih dari seluruh fitur hasil preprocessing.

Temuan Utama:
1. Logistic Regression tetap menjadi salah satu model terkuat dengan F2-score test turun sangat sedikit dari 0.7112 (baseline) ke 0.7078 (RFE). Hal ini menunjukkan bahwa pengurangan fitur tidak berdampak signifikan pada model linear ini.

2. XGBoost dan LightGBM mempertahankan performa tinggi, masing-masing dengan F2-score test turun sangat sedikit:

    - XGBoost: 0.6507 → 0.6598

    - LightGBM: 0.7017 → 0.6914. Ini menunjukkan bahwa model boosting cukup robust terhadap seleksi fitur.

3. Model berbasis pohon sederhana seperti Decision Tree dan Random Forest menunjukkan sedikit penurunan performa, mengindikasikan bahwa beberapa fitur yang dieliminasi mungkin masih penting bagi model tersebut.

  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">7.5 | Resampling (Balancing Data)</h2>
</div>

<div style='width: 100%; height: 0.05px; background-color: #BDC3C7; border-radius: 10px; margin-top: 20px; margin-bottom: 20px;'></div>

Eksperimen ini bertujuan untuk mengatasi ketidakseimbangan data (class imbalance) dengan menggunakan teknik resampling. Dua metode yang akan diuji adalah:

- SMOTE (Synthetic Minority Over-sampling Technique): Membuat sampel sintetis dari kelas minoritas dengan interpolasi.

- ADASYN (Adaptive Synthetic Sampling): Mirip SMOTE, namun menekankan pada sampel yang lebih sulit diklasifikasikan.

Kedua teknik akan diuji dengan model terbaik sejauh ini, yaitu Logistic Regression dan LightGBM, berdasarkan eksperimen 6.1 hingga 6.4.

```python
# Setup Cross Validation dan Scoring
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
f2_scorer = make_scorer(fbeta_score, beta=2)

# Model terbaik sejauh ini
models = {
    "Logistic Regression": LogisticRegression(random_state=42, max_iter=1000),
    "LightGBM": LGBMClassifier(random_state=42)
}

# Resamplers
resamplers = {
    "SMOTE": SMOTE(random_state=42),
    "ADASYN": ADASYN(random_state=42)
}

# Simpan hasil
results_resampling = []

# ➤ Baseline (Tanpa Resampling)
for model_name, model in models.items():
    print(f"Evaluating {model_name} without resampling (Baseline)...")

    f2_scores = cross_val_score(model, X_train_df, y_train, cv=kf, scoring=f2_scorer)
    mean_f2 = np.mean(f2_scores)
    std_f2 = np.std(f2_scores)

    model.fit(X_train_df, y_train)
    y_pred_test = model.predict(X_test_df)
    f2_test = fbeta_score(y_test, y_pred_test, beta=2)

    results_resampling.append({
        "Model": model_name,
        "Resampler": "No Resampling",
        "F2 Train Mean": round(mean_f2, 4),
        "F2 Train Std": round(std_f2, 4),
        "F2 Test": round(f2_test, 4)
    })

# ➤ SMOTE & ADASYN
for resampler_name, resampler in resamplers.items():
    for model_name, model in models.items():
        print(f"Evaluating {model_name} with {resampler_name}...")

        pipe = ImbPipeline([
            ('resample', resampler),
            ('model', model)
        ])

        f2_scores = cross_val_score(pipe, X_train_df, y_train, cv=kf, scoring=f2_scorer)
        mean_f2 = np.mean(f2_scores)
        std_f2 = np.std(f2_scores)

        pipe.fit(X_train_df, y_train)
        y_pred_test = pipe.predict(X_test_df)
        f2_test = fbeta_score(y_test, y_pred_test, beta=2)

        results_resampling.append({
            "Model": model_name,
            "Resampler": resampler_name,
            "F2 Train Mean": round(mean_f2, 4),
            "F2 Train Std": round(std_f2, 4),
            "F2 Test": round(f2_test, 4)
        })

# ➤ Tampilkan Hasil
df_resampling = pd.DataFrame(results_resampling)
display(df_resampling)
```

Penerapan teknik resampling seperti SMOTE dan ADASYN terbukti memberikan peningkatan performa yang signifikan terhadap kemampuan model dalam memprediksi pelanggan churn (kelas minoritas). Ini terlihat jelas terutama pada model Logistic Regression.

1. Logistic Regression mengalami peningkatan drastis dari baseline F2-score 0.5540 (tanpa resampling) menjadi:

    - 0.7122 dengan SMOTE

    - 0.7253 dengan ADASYN Hal ini menunjukkan bahwa model linear seperti Logistic Regression sangat terbantu oleh distribusi data yang lebih seimbang.

2. LightGBM, meskipun merupakan model yang kompleks dan powerful, justru menunjukkan sedikit penurunan performa ketika diberikan data hasil resampling:

    Dari baseline F2-score 0.5574 menjadi:

    - 0.6026 (SMOTE)

    - 0.5950 (ADASYN) Ini mengindikasikan bahwa model boosting seperti LightGBM mungkin sudah cukup baik menangani imbalance data tanpa perlu oversampling eksplisit, atau bahwa oversampling justru menambah noise.



**Kesimpulan Utama**

- Resampling terbukti krusial untuk model sederhana (linear/tree), namun perlu dievaluasi hati-hati pada model kompleks (boosting).

  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">8.2 | Definisi Parameter Space untuk RandomizedSearchCV</h2>

Pada bagian ini, kita akan melakukan hyperparameter tuning pertama menggunakan RandomizedSearchCV.​

Tujuannya adalah untuk menemukan kombinasi hyperparameter terbaik untuk model Logistic Regression + ADASYN, dengan fokus pada peningkatan performa metrik F2-score.


Berdasarkan [dokumentasi resmi scikit-learn](https://stackoverflow.com/questions/67513075/what-is-c-parameter-in-sklearn-logistic-regression?utm_source=chatgpt.com) , berikut adalah hyperparameter yang akan kita eksplorasi:​

| **Hyperparameter** | **Deskripsi**                                                                                   | **Range yang Dicoba**             |
|--------------------|-------------------------------------------------------------------------------------------------|-----------------------------------|
| `penalty`          | Jenis penalti regulasi untuk mencegah overfitting.                                              | `['l1', 'l2']`                    |
| `C`                | Invers dari kekuatan regularisasi; nilai kecil = regulasi lebih kuat.                          | `np.logspace(-3, 2, 10)`          |
| `solver`           | Algoritma optimasi untuk fitting model.                                                         | `['liblinear', 'saga']`           |

Langkah selanjutnya: Fine-tuning dengan **RandomSearchCV** untuk pencarian kombinasi optimal.


<div style="text-align: left;">
  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">8.3 | Implementasi RandomizedSearchCV</h2>
</div>

<div style='width: 100%; height: 0.05px; background-color: #BDC3C7; border-radius: 10px; margin-top: 20px; margin-bottom: 20px;'></div>

Pada subbab ini, dilakukan tuning hyperparameter untuk model Logistic Regression menggunakan RandomizedSearchCV. Tuning dilakukan terhadap parameter penalti (l1, l2), regularisasi (C), dan solver (liblinear, saga) untuk meningkatkan F2-score pada data training.

```python
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import make_scorer, fbeta_score
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import ADASYN
from scipy.stats import loguniform
import numpy as np

# Pipeline: ADASYN + Logistic Regression
pipe = ImbPipeline([
    ('resample', ADASYN(random_state=42)),
    ('model', LogisticRegression(random_state=42, max_iter=1000))
])

# Ruang parameter yang akan dituning
param_dist = {
    'model__penalty': ['l1', 'l2'],
    'model__C': loguniform(1e-3, 1e2),
    'model__solver': ['liblinear', 'saga']
}

# Cross-validation setup
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
f2_scorer = make_scorer(fbeta_score, beta=2)

# RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=param_dist,
    n_iter=20,  # kamu bisa ubah ke 30-50 untuk eksplorasi lebih dalam
    scoring=f2_scorer,
    n_jobs=-1,
    cv=kf,
    random_state=42,
    verbose=1
)

# Fit Random Search ke data training
random_search.fit(X_train_df, y_train)

# Tampilkan hasil tuning
print("Best Parameters:")
print(random_search.best_params_)

print("\nBest CV F2-score:")
print(round(random_search.best_score_, 4))
```

Tuning berhasil meningkatkan performa model dengan kombinasi optimal:

- penalty = 'l1'

- C ≈ 0.19

- solver = 'saga'

Ini adalah evaluasi akhir yang akan menunjukkan bagaimana model bekerja di data yang benar-benar belum pernah dilihat sebelumnya.

Tujuan Evaluasi:

2. Menghitung metrik F2-score pada test set

```python
from sklearn.metrics import classification_report, confusion_matrix, fbeta_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Ambil best estimator dari hasil random search
best_model = random_search.best_estimator_

# Prediksi terhadap test set
y_pred_test = best_model.predict(X_test_df)

# Hitung F2-score
f2_test_score = fbeta_score(y_test, y_pred_test, beta=2)
print(f"Final F2-score on Test Set: {round(f2_test_score, 4)}")

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_test, target_names=["Not Churn", "Churn"]))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred_test)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Churn", "Churn"])
disp.plot(cmap="Blues")
plt.title("Confusion Matrix: Final Tuned Model")
plt.show()
```

Secara default, scikit-learn menggunakan threshold 0.5. Namun, dalam kasus bisnis seperti churn prediction, threshold tersebut belum tentu optimal secara bisnis.

Kenapa Threshold Penting?
- Dengan menurunkan threshold, model cenderung menangkap lebih banyak churn (↑ recall), namun berisiko menaikkan FP (↓ precision).

- Dengan menaikkan threshold, model akan lebih selektif, tetapi bisa kehilangan pelanggan churn (↑ precision, ↓ recall).

- Di sinilah pentingnya analisis trade-off antara retention cost (FP) dan churn loss (FN).

**Analisis Trade-off**

Kita akan menggunakan pendekatan berikut:

- Plot Precision vs Recall untuk berbagai threshold

- Plot F2-score untuk memilih threshold optimal berbasis bisnis (Recall-weighted)

- Menandai titik-titik potensial yang bisa digunakan oleh tim marketing berdasarkan toleransi biaya FP vs risiko FN

### **Threshold Curve & F2-Score Optimization**
---

Menentukan nilai threshold probabilitas optimal berdasarkan keseimbangan antara Recall dan Precision, dengan mempertimbangkan F2-Score sebagai metrik utama (karena recall lebih penting dari precision).


Langkah Evaluasi:
1. Gunakan model final: **Logistic Regression + ADASYN (default hyperparameters)**


2. Hitung probabilitas prediksi (predict_proba) terhadap X_test_df.

3. Evaluasi berbagai nilai threshold dari 0.1 hingga 0.9.

4. Hitung precision, recall, dan F2-score untuk tiap threshold.

5. Visualisasikan hasilnya dalam bentuk garis kurva.

```python
from sklearn.metrics import precision_score, recall_score, fbeta_score
import matplotlib.pyplot as plt
import numpy as np

# 1. Prediksi probabilitas churn dari model final
y_proba = final_model.predict_proba(X_test_df)[:, 1]

# 2. Siapkan threshold range dan hasil evaluasi
thresholds = np.arange(0.1, 0.91, 0.01)
f2_scores, precisions, recalls = [], [], []

# 3. Loop untuk evaluasi tiap threshold
for t in thresholds:
    y_pred = (y_proba >= t).astype(int)
    f2_scores.append(fbeta_score(y_test, y_pred, beta=2))
    precisions.append(precision_score(y_test, y_pred))
    recalls.append(recall_score(y_test, y_pred))

# 4. Visualisasi trade-off
plt.figure(figsize=(10, 6))
plt.plot(thresholds, f2_scores, label="F2 Score", color="blue", linewidth=2)
plt.plot(thresholds, precisions, label="Precision", linestyle='--', color="orange")
plt.plot(thresholds, recalls, label="Recall", linestyle='--', color="green")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.title("Threshold vs F2 / Precision / Recall")
plt.legend()
plt.grid(True)
plt.show()

# 5. Temukan threshold terbaik
best_threshold = thresholds[np.argmax(f2_scores)]
best_f2 = max(f2_scores)

print(f"Optimal Threshold: {best_threshold:.2f}")
print(f"Best F2-score at Optimal Threshold: {best_f2:.4f}")
```

Dari visualisasi dan evaluasi terhadap berbagai nilai threshold, diperoleh bahwa:

- Threshold optimal adalah 0.43, yang menghasilkan F2-score tertinggi sebesar 0.7625.

- Recall tetap tinggi (~0.83) pada threshold ini, yang sangat penting dalam konteks churn prediction.

- Precision meningkat signifikan dibanding threshold default (0.5), tanpa mengorbankan terlalu banyak recall.

**Kenapa Threshold 0.43 Ideal?**

Dalam konteks bisnis IndoHome, yang menjadi perhatian utama adalah:

- False Negative (FN) = Pelanggan churn yang gagal terdeteksi → langsung berdampak pada kehilangan pendapatan.

- Maka, kita butuh model yang sensitif terhadap churn → recall tinggi lebih diprioritaskan.

- Namun, precision juga penting untuk efisiensi biaya promosi agar tidak banyak pelanggan loyal yang salah diprediksi churn.

Threshold 0.43 memberikan keseimbangan terbaik antara:

- Recall tinggi (menjaga potensi pendapatan)

- Precision cukup (efisiensi biaya retensi)



**Business Implication:**

Dengan threshold ini:

- Model akan mengidentifikasi lebih banyak pelanggan churn.

- Tim pemasaran bisa lebih tepat sasaran saat memberikan diskon atau program loyalitas.

- Secara keseluruhan, pendekatan ini mengurangi profit loss dan operation loss seperti yang dijelaskan di Chapter 1.5 | Metric Evaluation.

```python
import joblib
import json
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import ADASYN

# 1. Definisikan fitur
numerical = ['tenure', 'MonthlyCharges', 'TotalCharges', 'TotalServices']
categorical = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents',
    'PhoneService', 'MultipleLines',
    'InternetService', 'OnlineSecurity', 'OnlineBackup',
    'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaperlessBilling',
    'PaymentMethod', 'TenureGroup'
]

# 2. Preprocessor
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numerical),
    ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical)
])

# 3. Pipeline Produksi Lengkap
final_pipeline = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('resample', ADASYN(random_state=42)),
    ('model', LogisticRegression(max_iter=1000, random_state=42))
])

# 4. Fit ke data training mentah
final_pipeline.fit(X_train_df, y_train)

# # 5. Simpan pipeline model produksi
# joblib.dump(final_pipeline, 'deployment/final_model_pipeline.pkl')

# # 6. Simpan threshold optimal
# with open('deployment/threshold.json', 'w') as f:
#     json.dump({'threshold': 0.43}, f)

# # 7. Simpan daftar fitur (metadata)
# feature_list = {
#     'numerical': numerical,
#     'categorical': categorical
# }
# with open('deployment/final_feature_list.json', 'w') as f:
#     json.dump(feature_list, f, indent=2)

# # 8. Simpan metadata versi pipeline
# with open('deployment/pipeline_version.txt', 'w') as f:
#     f.write("Model: Logistic Regression + ADASYN\n")
#     f.write("Preprocessing: StandardScaler & OneHotEncoder\n")
#     f.write("Threshold: 0.43\n")
#     f.write("Version: 1.0\n")

print("✅ Final pipeline produksi berhasil disimpan.")
```

  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">10.2 | Koefisien Logistic Regression</h2>

<div style='width: 100%; height: 0.05px; background-color: #BDC3C7; border-radius: 10px; margin-top: 20px; margin-bottom: 20px;'></div>

Pada subbab ini, kita akan melihat langsung koefisien regresi (log odds) dari model Logistic Regression yang telah dilatih. Nilai koefisien menunjukkan:

1. Arah pengaruh fitur terhadap prediksi churn:

    - Koefisien positif → meningkatkan peluang pelanggan churn.

    - Koefisien negatif → menurunkan peluang churn.

2. Besarnya dampak: Semakin besar absolut koefisien, semakin kuat pengaruh fitur tersebut terhadap keputusan model.

```python
# Mengambil model dari pipeline
logreg_model = pipeline.named_steps['model']
ohe = pipeline.named_steps['preprocessor'].named_transformers_['cat']
encoded_cat_names = ohe.get_feature_names_out(categorical)

# Menggabungkan nama fitur numerik dan kategorikal (setelah one-hot)
feature_names = numerical + list(encoded_cat_names)

# Mengambil koefisien model
coefficients = logreg_model.coef_[0]

# Membuat dataframe koefisien
df_coef = pd.DataFrame({
    'Feature': feature_names,
    'Coef (Log Odds)': coefficients
}).sort_values(by='Coef (Log Odds)', ascending=False)

# Menampilkan 10 pengaruh positif & negatif terbesar
df_top = pd.concat([
    df_coef.head(10),
    df_coef.tail(10)
])

df_top.reset_index(drop=True, inplace=True)
df_top
```

Interpretasi Awal
- Fitur dengan nilai koefisien tertinggi memberikan kontribusi terbesar untuk meningkatkan probabilitas churn.

- Sebaliknya, fitur dengan nilai koefisien negatif terbesar cenderung berperan sebagai faktor penahan churn — pelanggan dengan karakteristik ini cenderung lebih loyal.

Pada subbab berikutnya (10.3), koefisien ini akan kita ubah menjadi odds ratio agar lebih mudah dipahami secara bisnis.

  <h2 id="context" style="background-color: #E60000; color: white; padding: 10px 20px; margin-bottom: 20px; border-radius: 8px; font-weight: bold; display: inline-block;">10.3 | Odds Ratio & Interpretasi Fitur Penting</h2>
</div>

<div style='width: 100%; height: 0.05px; background-color: #BDC3C7; border-radius: 10px; margin-top: 20px; margin-bottom: 20px;'></div>

Transformasi koefisien logistik menjadi odds ratio dilakukan untuk mempermudah pemahaman bisnis terhadap pengaruh masing-masing fitur. Odds ratio menjelaskan seberapa besar peluang churn berubah jika fitur tertentu naik satu unit, dengan asumsi fitur lainnya tetap (ceteris paribus).

```python
# Tambahkan kolom Odds Ratio
df_top['Odds Ratio'] = np.exp(df_top['Coef (Log Odds)']).round(3)

# Urutkan kembali berdasarkan odds tertinggi dan terendah
df_top_sorted = df_top.sort_values(by='Odds Ratio', ascending=False)
df_top_sorted.reset_index(drop=True, inplace=True)

df_top_sorted
```

Fitur Pendorong Churn (Odds Ratio > 1):

- Fiber optic, Electronic check, StreamingMovies_Yes, PaperlessBilling_Yes

- Ini adalah fitur prioritas intervensi bisnis: pelanggan yang memakai layanan-layanan ini atau metode pembayaran ini perlu dipantau lebih ketat.

Fitur Pelindung (Odds Ratio < 1):

- Contract_Two year, TechSupport_Yes, OnlineSecurity_Yes, TenureGroup_Lama, TotalServices

- Ini adalah profil pelanggan loyal — bisa menjadi baseline untuk strategi retensi atau upsell.

1. Tujuan Monitoring & Retraining

- **Menjaga akurasi model** dalam menghadapi perubahan perilaku pelanggan (drift).
- **Memastikan threshold klasifikasi tetap optimal** untuk tujuan bisnis (F2-score & recall).
- **Mengantisipasi data distribution shift** (misalnya saat ada promosi besar-besaran, perubahan paket layanan, atau musim pendaftaran baru).

3. Rencana Monitoring Berkala

| Komponen           | Frekuensi       | Metode/Tools                           | Tindak Lanjut                                     |
|--------------------|------------------|----------------------------------------|--------------------------------------------------|
| **F2-score & Recall**  | Bulanan          | Evaluasi via confusion matrix dari prediksi mingguan | Jika F2-score < baseline → lakukan review model |
| **Threshold**       | 3–6 bulan sekali | Uji ulang berdasarkan PR-curve terbaru | Update nilai di `threshold.json` jika perlu      |
| **Distribusi Fitur** | Bulanan          | Perbandingan histogram fitur baru vs. training | Jika ada pergeseran signifikan → flag retraining |
| **Feedback dari CRM** | Mingguan        | Analisis manual dari tim marketing      | Catat false negative atau false positive yang tinggi |

4. Rencana Retraining Model

Kapan model perlu retraining?
- Penurunan F2-score atau recall selama **2 bulan berturut-turut**
- Muncul **fitur baru penting** (mis. layanan baru, paket promo)
- Perubahan regulasi atau pricing strategi besar
- **Drift data** signifikan terdeteksi (dengan alat seperti DataDrift Monitor, EvidentlyAI, atau analisis manual)

Alur Retraining yang Direkomendasikan:
1. **Ambil data terbaru** (3–6 bulan terakhir)
2. **Gabungkan dengan data lama**, lakukan balancing ulang jika perlu (ADASYN)
3. **Evaluasi model baru** terhadap metrik bisnis (recall, F2-score, false negative)
4. **Simpan versi baru** di `deployment/` dan update `pipeline_version.txt`
5. **Uji di staging environment**, sebelum digunakan live oleh CRM/marketing

5. Tracking & Audit Versi Model

Berdasarkan seluruh proses analisis dan hasil implementasi model churn, berikut adalah tiga pilar rekomendasi utama untuk mendorong keberlanjutan dan dampak maksimal proyek ini:


#### **1. Untuk Bisnis IndoHome**

----

1. Integrasikan model churn ke sistem CRM operasional, yaitu menghubungkan pipeline prediksi churn ke data pelanggan IndoHome secara otomatis dan berkesinambungan, agar hasil prediksi dapat langsung digunakan oleh tim marketing dan CRM untuk kampanye retensi harian/mingguan.
**Catatan**: Saat ini model telah dideploy dalam bentuk aplikasi Streamlit, yang dapat digunakan untuk keperluan interaktif. Namun untuk adopsi skala besar, dibutuhkan integrasi lebih lanjut ke dalam sistem CRM agar prediksi dilakukan secara otomatis dan dipakai langsung dalam pipeline bisnis IndoHome.

2. Terapkan strategi intervensi tersegmentasi berdasarkan skor churn:
   - Skor > 0.80 → Penawaran spesial, diskon retensi, atau intervensi prioritas (call center).
   - Skor 0.60–0.80 → Loyalty program, bundling, atau konten edukatif.
   - Skor < 0.43 → Monitoring pasif, pelanggan loyal tidak perlu intervensi khusus.
3. Gunakan hasil interpretasi model (Chapter 10) untuk menyusun strategi retensi berbasis fitur:

- Fitur yang **meningkatkan risiko churn**:
  - `InternetService_Fiber optic`
  - `MultipleLines_Yes`
  - `PaymentMethod_Electronic check`
  - `StreamingTV_Yes`
- Fitur yang **menurunkan risiko churn (pelanggan loyal)**:
  - `Contract_Two year`
  - `OnlineSecurity_Yes`
  - `TechSupport_Yes`
  - `TenureGroup_Lama`

**Strategi tindakan**:
- Konversi pelanggan kontrak bulanan ke tahunan.
- Tambahkan atau promosikan fitur layanan seperti `Tech Support`, `Online Security`.
- Fokus intervensi pada pelanggan fiber optic dan pengguna e-check.
- Bangun program loyalti khusus untuk pelanggan dengan profil loyal, misalnya pelanggan `TenureGroup_Lama`.

Dengan menggabungkan prediksi dan interpretasi model, strategi retensi tidak lagi bersifat asumtif, melainkan data-driven, dapat dijelaskan, dan langsung dapat ditindaklanjuti.

#### **2. Untuk Model Machine Learning**

---

1. Pantau performa model (F2-score, Recall) secara berkala setiap bulan. Gunakan data aktual hasil kampanye untuk validasi prediksi.
2. Review threshold dan distribusi prediksi setiap 3–6 bulan untuk memastikan cutoff tetap optimal terhadap dinamika pasar dan pelanggan.
3. Lakukan retraining model secara penuh setiap 6–12 bulan, atau ketika terjadi:
   - Penurunan performa metrik klasifikasi
   - Perubahan produk/paket layanan baru
   - Munculnya fitur baru dari CRM atau kebijakan internal

#### **3. Untuk Dataset & Infrastruktur**

---

1. Perkaya dataset dengan variabel tambahan, seperti:
   - Riwayat interaksi call center
   - Feedback kepuasan pelanggan
   - Event marketing atau promo yang pernah diikuti pelanggan
2. Bangun pipeline integrasi data otomatis dari sistem CRM ke engine prediksi untuk meningkatkan automasi.
3. Catat log histori prediksi dan feedback, agar dapat digunakan sebagai data baru untuk retraining model dan evaluasi efektivitas model secara longitudinal.

Dengan pendekatan berbasis data yang dapat dijelaskan (explainable ML), solusi ini mendukung strategi CMO IndoHome secara langsung dalam meningkatkan retensi pelanggan, mengurangi biaya kampanye yang tidak efisien, dan memperkuat kepercayaan stakeholder terhadap teknologi AI dalam pengambilan keputusan pemasaran.

Solusi ini bukan hanya *prediksi churn*, tapi fondasi sistem retensi pelanggan berbasis machine learning yang adaptif dan berorientasi dampak nyata.
